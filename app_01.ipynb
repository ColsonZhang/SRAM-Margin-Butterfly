```python
x = np.linspace(-0.1, 1.2, round(1.2/0.01))
offset_y = 0
y = x + offset_y

x_s = np.linspace(-0.1, 1.2, round(1.2/0.01))
offset_y_s = 0.8
y_s = x_s + offset_y_s

x_e = np.linspace(-0.1, 1.2, round(1.2/0.01))
offset_y_e = -0.4
y_e = x_e + offset_y_e
```

```python
cross_1 =  intersection(x,y,curve_1[0][0],curve_1[0][1])
cross_2 =  intersection(x,y,curve_2[0][0],curve_2[0][1])
```

```python
fig,axs = plt.subplots()
axs.plot(curve_1[0][0],curve_1[0][1])
axs.plot(curve_2[0][0],curve_2[0][1])

axs.plot(x,y)
axs.plot(x_s,y_s, ":")
axs.plot(x_e,y_e, ":")


axs.plot([cross_1[0],cross_2[0]], [cross_1[1], cross_1[1]], "c--")
axs.plot([cross_2[0],cross_2[0]], [cross_1[1], cross_2[1]], "c--")
axs.plot([cross_2[0],cross_1[0]], [cross_2[1], cross_2[1]], "c--")
axs.plot([cross_1[0],cross_1[0]], [cross_2[1], cross_1[1]], "c--")

axs.scatter(cross_1[0], cross_1[1],c='r')
axs.scatter(cross_2[0], cross_2[1],c='r')

axs.annotate('line-scan',xy=(0.4,1.2),xytext=(1.0,0.6),arrowprops=dict(arrowstyle="<-",connectionstyle="arc3"))

plt.axis('scaled')
plt.xlim(-0.2,1.4)
plt.ylim(-0.2,1.4)

plt.tight_layout()
plt.show()
```
